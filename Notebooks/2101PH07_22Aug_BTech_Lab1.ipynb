```python
import numpy as np
```

# Example Circuit

```python
from qiskit import QuantumCircuit, QuantumRegister

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.measure_all()
qc.draw('mpl')
```

# Exercise 1

```python
qc = QuantumCircuit(3)
qc.x(0)
qc.h(1)
qc.z(2)
qc.crx(np.pi/4,0,1)
qc.mcry(-np.pi/8,q_controls=[0, 1], q_target=2)
qc.draw('mpl')
```

```python
qc.width()
```

```python
qc.depth()
```

# Quantum Information Class

```python
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
sv = Statevector.from_instruction(qc)
print(sv)
sv.draw('latex')
```

# Exercise 2

```python
qc = QuantumCircuit(4)
qc.x(0)
qc.z(1)
qc.barrier()
qc.cx(1,2)
qc.crz(np.pi/12,0,3)
qc.rx(-np.pi/7,2)
qc.draw('mpl')
```

```python
sv = Statevector.from_instruction(qc)
print(sv)
sv.draw('latex')
```

# Evolve Method

```python
qc2 = QuantumCircuit(4)
qc2.ry(np.pi/8,3)
qc2.x(2)
qc2.draw('mpl')
```

```python
sv2 = sv.evolve(qc2)
print(sv2)
sv2.draw('latex')
```

# Visualization

```python
from qiskit.visualization import plot_state_city

plot_state_city(sv2)
```

# Measurement

```python
from qiskit.visualization import plot_histogram

prob = sv2.probabilities_dict()
plot_histogram(prob)
```
