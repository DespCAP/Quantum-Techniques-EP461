# Universal Gate

```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
import numpy as np
import qiskit
from qiskit.circuit.library import RYGate
from qiskit.circuit import ControlledGate
```

```python
t0 = np.pi/6
t1 = -np.pi/8
t2 = np.pi/2
t3 = np.pi/7
```

## Using CXX Gates

```python
qc0 = QuantumCircuit(3)
qc0.x(0)
qc0.x(1)
qc0.mcry(t0,[0,1],2)
qc0.x(0)
qc0.x(1)
qc0.barrier()
qc0.x(0)
qc0.mcry(t1,[0,1],2)
qc0.x(0)
qc0.barrier()
qc0.x(1)
qc0.mcry(t2,[0,1],2)
qc0.x(1)
qc0.barrier()
qc0.mcry(t3,[0,1],2)
qc0.draw('mpl')
```

```python
sv0 = np.round(Statevector.from_instruction(qc0).data, 5)
print(sv0)
```

## Using Equivalent circuit

```python
a0 = (t0 + t1 + t2 + t3)/4
a1 = (t0 - t1 + t2 - t3)/4
a2 = (t0 - t1 - t2 + t3)/4
a3 = (t0 + t1 - t2 - t3)/4
```

```python
qc1 = QuantumCircuit(3)
qc1.ry(a0, 2)
qc1.cx(1, 2)
qc1.ry(a1, 2)
qc1.cx(0, 2)
qc1.ry(a2, 2)
qc1.cx(1, 2)
qc1.ry(a3, 2)
qc1.cx(0, 2)
qc1.draw(output='mpl')
```

```python
sv1 = np.round(Statevector.from_instruction(qc1).data, 5)
print(sv1)
```
