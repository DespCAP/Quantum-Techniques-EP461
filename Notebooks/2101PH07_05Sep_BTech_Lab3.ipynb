# Two Qubit Gates

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
import numpy as np
from qiskit.visualization import plot_state_city
```

## CX Gate

```python
qc0 = QuantumCircuit(2)
qc0.cx(1,0)
unitary0 = Operator(qc0)
print(unitary0)
qc0.draw('mpl')
```

```python
qc1 = QuantumCircuit(2)
qc1.h([0,1])
qc1.cx(0, 1)
qc1.h([0,1])
unitary1 = Operator(qc1)
print(unitary1)
qc1.draw('mpl')
```

## CZ Gate

```python
qc2 = QuantumCircuit(2)
qc2.cz(0, 1)
unitary2 = Operator(qc2)
print(unitary2)
qc2.draw('mpl')
```

```python
qc3 = QuantumCircuit(2)
qc3.cz(1, 0)
unitary3 = Operator(qc3)
print(unitary3)
qc3.draw('mpl')
```

## SWAP Gate

```python
qc4 = QuantumCircuit(2)
# a
qc4.ry(np.pi/6, 0)
qc4.x(1)
qc4.ry(-np.pi/8, 1)
qc4.cx(0, 1)
qc4.cx(1, 0)
qc4.cx(0, 1)
sv_after_swap = Statevector.from_instruction(qc4)
print("After swap", sv_after_swap)
qc4.draw('mpl')
```

```python
plot_state_city(sv_after_swap)
```

```python
qc5 = QuantumCircuit(2)
# a
qc5.ry(np.pi/6, 1)
qc5.x(0)
qc5.ry(-np.pi/8, 0)
sv_before_swap = Statevector.from_instruction(qc5)
print("Swapped states", sv_before_swap)
qc5.draw('mpl')
```

```python
plot_state_city(sv_before_swap)
```

## Quantum Multiplexor

```python
theta_0 = np.pi/12
theta_1 = np.pi/6
```

```python
qc6 = QuantumCircuit(2)
qc6.cry(theta_0,0,1,ctrl_state=0)
qc6.cry(theta_1,0,1)
unitary6 = Operator(qc6)
print(unitary6)
qc6.draw('mpl')
```

```python
qc7 = QuantumCircuit(2)
qc7.ry((theta_0+theta_1)/2,1)
qc7.cx(0,1)
qc7.ry((theta_0-theta_1)/2,1)
qc7.cx(0,1)
unitary7 = Operator(qc7)
print(unitary7)
qc7.draw('mpl')
```

## RY-RZ Decomposition of RX Gate

```python
qc8 = QuantumCircuit(1)
qc8.rx(np.pi/3,0)
unitary8 = Operator(qc8)
print(unitary8)
qc8.draw('mpl')
```

```python
qc9 = QuantumCircuit(1)
qc9.rz(-np.pi/2,0)
qc9.ry(np.pi/3,0)
qc9.rz(np.pi/2,0)
unitary9 = Operator(qc9)
print(unitary9)
qc9.draw('mpl')
```
